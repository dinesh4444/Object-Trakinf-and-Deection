```python
import cv2
import imutils
```

```python
# HSV values for color calibration
redlower = (87,93,203)
redupper = (177,255,255)
```

```python
#intializing Camera
cam = cv2.VideoCapture(0)

#infinite loop

while True:
    #Reading camrea frame
    (grabbed,frame) = cam.read()
    #resizing image
    frame = imutils.resize(frame, width=600)
    #doing smoothening
    blurred = cv2.GaussianBlur(frame, (11,11), 0)
    #convering BGR to HSV
    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)
    #for detecting specified color for that we have to mask the color
    mask = cv2.inRange(hsv, redlower, redupper)
    #erode=remove the holes and make thin
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations = 2)
    #finding contours to track object and identify boundries of object
    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]
    center = None
    if len(cnts) > 0:
        #cv2.contourArea find the max contour area
        c = max(cnts, key=cv2.contourArea)
        #drawing min enclosure circle
        ((x,y),redius) = cv2.minEnclosingCircle(c)
        M = cv2.moments(c)
        #center of the circle which will drawn
        center = (int(M["m10"] / M["m00"]), int(M["m01"] / M["m00"]))
        #confition for drawing circle
        if redius > 10:
            #drawing circle arround the frame
            cv2.circle(frame, (int(x), int(y)), int(redius), (0,255,255),2)
            #ploting the centroid of the circle with dot 
            cv2.circle(frame, center, 5, (0,0,255), -1)
            #condition for redius
            if redius > 250:
                print('Stop')
            else:
                if (center[0] < 150):
                    print('Left')
                elif (center[0] > 450):
                    print("Right")
                elif (center[0] < 250):
                    print("Front")
                else:
                    print("Stop")
    cv2.imshow("Frame", frame)
    key=cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
        
cam.release()
cv2.destroyAllWindows()
```
